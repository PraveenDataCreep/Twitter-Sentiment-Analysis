```
import spacy
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
import pandas as pd
spark = SparkSession \
        .builder \
        .appName("Twitter Data Loading") \
        .config("spark.gars.config.option", "gars-config") \
        .getOrCreate()
spark.conf.set("spark.sql.execution.arrow.enabled", "true")
glove = '/dbfs/ml/twitter_embeddings/spacy_glove_model'
```

```
df = spark.read.table('glove_embeddings_stanford')
```

```
display(df.limit(10))
```

```
display(df.agg( *[ count(c).alias(c) for c in df.columns ] ))
```

```
labelled = spark.read.table('clean_labelled_data')
display(labelled.agg( *[ count(c).alias(c) for c in labelled.columns]))
```

```
df = df.dropna(how='any')
display(df.agg( *[ count(c).alias(c) for c in df.columns ] ))
```

```
(trainingData, testData) = df.randomSplit([0.7,0.3], seed=42)
trainingData.cache()
testData.cache()
print (trainingData.count())
print (testData.count())
```

```
from pyspark.ml.feature import VectorAssembler
from pyspark.ml import Pipeline
from pyspark.mllib.tree import GradientBoostedTrees, GradientBoostedTreesModel
from pyspark.ml.classification import GBTClassifier
from pyspark.mllib.util import MLUtils
from pyspark.ml.feature import StringIndexer

labelIndexer = StringIndexer(inputCol="sentiment", outputCol="label")
inputCol = [f'vectors_feat[{x}]' for x in range(200)]
vecAssembler = VectorAssembler(inputCols = inputCol, outputCol='features')
gbt = GBTClassifier(labelCol="label", featuresCol="features", maxIter=200,  stepSize=0.08, subsamplingRate=0.7)

# Chain indexers and GBT in a Pipeline
pipeline = Pipeline(stages=[labelIndexer, vecAssembler, gbt])

# Train model.  This also runs the indexers.
model = pipeline.fit(trainingData)
```

```
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
(cvData, testData) = testData.randomSplit([0.5,0.5], seed=42)
predictions = model.transform(cvData)
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction",
                                              metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print( "Model Accuracy: ", accuracy)
```

```
predictions = model.transform(cvData)
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction",
                                              metricName="f1")
accuracy = evaluator.evaluate(predictions)
print( "Model Accuracy: ", accuracy)
```

```
predictions = model.transform(testData)
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction",
                                              metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print( "Model Accuracy: ", accuracy)
```
